```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
sns.scatterplot(data=df, x="bmi", y="charges", hue="sex")
plt.show()
```

```python
sns.relplot(data=df, x="bmi", y="charges",col="sex",kind="scatter",size="region")
plt.show()
```

```python
sns.relplot(data=df, x="bmi", y="charges", col="region", kind="scatter", style = "sex",
           alpha = .5, hue = "smoker", palette = ['black', 'g'], col_wrap = 2)
plt.show()
```
