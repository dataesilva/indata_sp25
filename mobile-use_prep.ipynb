```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
mb = pd.read_csv('mobile-use.csv')
mb.head()
```

```python
sns.scatterplot(mb, x = 'Age', y = 'spComfort', alpha = .2)

plt.show()
```

```python
sns.scatterplot(mb, x = 'Age', y = 'spComfort', hue = 'Sex', alpha = .2)

plt.show()
```

```python
sns.stripplot(mb, x = 'Sex', y = 'spComfort', alpha = .2)

plt.show()
```

```python
sns.scatterplot(mb[mb['Sex'] == 1], x = 'Age', y = 'spComfort')
sns.scatterplot(mb[mb['Sex'] == 2], x = 'Age', y = 'spComfort')
plt.show()
```

```python
sns.stripplot(mb, x = 'Education', y = 'spComfort')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.stripplot(mb, x = 'Ethnicity', y = 'spComfort', hue = 'Age')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.scatterplot(mb[mb['Ethnicity'] == 'White'], x = 'Age', y = 'spComfort', hue = 'Sex')
plt.show()
```

```python
sns.histplot(mb, x = 'spComfort')
```

```python
mb[mb['spComfort'] < 2]
```

```python
sns.stripplot(mb, x = 'spBrand', y = 'spComfort', alpha = .2)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.histplot(mb[mb['spBrand'] == 7], x = 'spComfort')
plt.show()
```

```python
sns.histplot(mb[mb['spBrand'] == 12], x = 'spComfort')
plt.show()
```

```python
g = sns.FacetGrid(mb, col="spBrand", col_wrap=3)
g.map(sns.histplot, 'spComfort')
plt.show()
```
