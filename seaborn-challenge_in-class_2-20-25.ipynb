```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
insurance = pd.read_csv('insurance.csv')
insurance.shape
```

```python
insurance.columns
```

```python
insurance.dtypes
```

```python
insurance.head()
```

```python
sns.relplot(data = insurance, x = 'bmi', y = 'charges', col = 'sex', kind = 'scatter')
# Worth adding
plt.show()
```

```python
# Discuss the value of using size for the region (categorical vs continuous)
sns.relplot(data = insurance,
            x = 'bmi', y = 'charges', col = 'sex',
            kind = 'scatter', size = 'region')
plt.show()
```

```python
insurance['smoker_ind'] = insurance['smoker'].map({'yes': 'Smoker', 'no':'Non-smoker'})
```

```python
insurance.head()
```

```python
sns.relplot(data = insurance,
           x = 'bmi', y = 'charges', col = 'sex', hue = 'smoker_ind', size = 'region', alpha = 0.5,
           kind = 'scatter')
plt.show()
```

```python
sns.pairplot(data = insurance, hue = 'smoker_ind')
plt.show()
```

```python
sns.pairplot(data = insurance, hue = 'smoker_ind', kind = 'reg')
plt.show()
```

```python
sns.pairplot(data = insurance, hue = 'region')
plt.show()
```

```python
sns.relplot(insurance, x = 'age', y = 'charges', style = 'smoker_ind', hue = 'sex',
           palette ="crest")
plt.show()
```

```python
sns.catplot(insurance, kind = 'bar', x = 'sex', y = 'charges', col = 'region')
plt.show()
```

```python
sns.catplot(insurance, kind = 'box', x = 'sex', y = 'charges', col = 'region', hue = 'sex',
            fill = False, saturation = .3)
plt.show()
```
