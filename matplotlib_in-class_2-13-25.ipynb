```python
import pandas as pd
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
```

```python
com = pd.read_csv('commute.csv')
com.shape
```

```python
com.head()
```

```python
com.columns
```

```python
len(com.index)
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(com['TripID'], com['CommuteTime'])


plt.show()
```

```python
fig2, axs2 = plt.subplots(5, 7)
axs2[2,2].plot(com['TripID'], com['CommuteTime'])

plt.tight_layout()
plt.show()
```

```python
fig3, axs3 = plt.subplots(1, 2, figsize = (8, 3))
axs3[1].set_xlabel('Trip ID')
axs3[0].set_ylabel('Commute Time')

plt.show()
```

```python
fig4, axs4 = plt.subplots(2, 3)
axs4[0,0].hist(com['CommuteTime'])
axs4[0,1].hist(com['CommuteTime'], bins = 3)

plt.tight_layout()
plt.show()
```

```python
w_counts = com.groupby('Weather').count()['TripID']
w_counts
```

```python
w_counts.index
```

```python
w_counts.values
```

```python
fig5, ax5 = plt.subplots()
ax5.bar(x = w_counts.index, height = w_counts.values)

plt.show()
```

```python
fig6, ax6 = plt.subplots()
ax6.scatter(com['TripID'], com['CommuteTime'])

plt.show()
```
