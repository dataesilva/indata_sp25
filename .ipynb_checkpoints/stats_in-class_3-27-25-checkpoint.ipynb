```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
ces = pd.read_csv('cces_example_data.csv')
ces.shape
```

```python
ces.head()
```

```python
ces['pid3']
```

```python
ces['sex']
```

```python
ces['pid3_c'] = ces['pid3'].astype('category')
ces['sex_c'] = ces['sex'].astype('category')
```

```python
ces['pid3_c'].value_counts()
```

```python
ces['sex_c'].value_counts()
```

```python
pid3_map = {1: 'Democratic', 2: 'Republican', 3: 'Independent'}
sex_map = {1: 'Male', 2: 'Female'}
```

```python
ces['pid3_c'] = ces['pid3_c'].cat.rename_categories(pid3_map)
ces['sex_c'] = ces['sex_c'].cat.rename_categories(sex_map)
```

```python
ces['sex_c'].value_counts()
```

```python
ces.columns
```

```python
fig1, axs1 = plt.subplots(1, 2)
axs1[0].hist(ces['InterGBiasParty'])
axs1[1].hist(ces['ideo5'])

plt.show()
```

```python
ces['ideo5'].value_counts()
```

```python
ces_fixed = ces[ces['ideo5'] != 6]
```

```python
ces_fixed.shape
```

```python
model1 = smf.ols("InterGBiasParty ~ ideo5", data = ces_fixed)
```

```python
fit1 = model1.fit()
fit1.summary()
```

```python
sns.jointplot(kind = 'reg', data = ces_fixed,
              x = 'ideo5', y = 'InterGBiasParty')
plt.show()
```

```python
sns.lmplot(data = ces_fixed, x = 'ideo5',
          y = 'InterGBiasParty',
          hue = 'pid3_c', x_jitter = .1,
          scatter_kws={'alpha':0.2})
plt.show()
```
