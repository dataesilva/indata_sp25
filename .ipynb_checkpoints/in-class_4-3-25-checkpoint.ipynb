```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
mb = pd.read_csv('mobile-use.csv')
mb.columns
```

```python
mb.shape
```

```python
sns.scatterplot(mb, x = 'Age', y = 'spComfort', alpha = .2)
plt.show()
```

```python
sns.scatterplot(mb, x = 'Age', y = 'spComfort', hue = 'Ethnicity', alpha = .2)
plt.show()
```

```python
g = sns.FacetGrid(mb, col = 'Ethnicity', col_wrap = 3)
g.map(sns.histplot, 'spComfort')
plt.show()
```

```python
sns.scatterplot(mb[mb['Age'] > 65], x = 'Age', y = 'spComfort', alpha = .2)
plt.show()
```

```python
sns.stripplot(mb, x = 'spBrand', y = 'spComfort')
plt.show()
sns.boxplot(mb, x = 'spBrand', y = 'spComfort')
plt.show()
```

```python
sns.lmplot(mb, y = 'spComfort', x = 'Age', ci = None, hue = 'spBrand')
plt.show()
```

```python
b = sns.FacetGrid(mb, col = 'spBrand', col_wrap = 4)
b.map_dataframe(sns.scatterplot, x = 'Age', y = 'spComfort')
plt.show()
```

```python
b = sns.FacetGrid(mb, col = 'spBrand', col_wrap = 4)
b.map_dataframe(sns.regplot, x = 'Age', y = 'spComfort', ci = None)
plt.show()
```

```python
sns.lmplot(mb[mb['spBrand'] == 14], y = 'spComfort', x = 'Age', hue = 'Sex')
plt.show()
```

```python
sns.lmplot(mb, y = 'spAge', x = 'spComfort')
plt.show()
```

```python
sns.stripplot(mb, x = 'Age', y = 'Education', alpha = .2)
plt.show()
```

```python
sns.stripplot(mb, y = 'spComfort', x = 'spBrand', alpha = .2, hue = 'Sex')
plt.show()
```
