```python
mb['Sex_c'] = mb['Sex'].astype('category')
```

```python
app3 = Dash()

app3.layout = html.Div([html.H1('Hello'),
                        html.H3('My App is great'),
                        dcc.Graph(id = 'spComfort-spAge_with-slider'),
                        dcc.Slider(
                            mb['Age'].min(),
                            mb['Age'].max(),
                            step = 1,
                            value = mb['Age'].min(),
                            id = 'Age_slider'
                        )])

@callback(
        Output('spComfort-spAge_with-slider', 'figure'),
        Input('Age_slider', 'value')
)
def update_figure(selected_age):
    fig = px.scatter(mb, x = 'spAge', y = 'spComfort', color = 'Sex_c')
    return fig

if __name__ == '__main__':
    app3.run(debug = True)
```

```python
app6 = Dash()

app6.layout = html.Div([html.H1('Hello'),
                        html.H3('My App is great'),
                        html.Div([
                        dcc.Graph(id = 'spComfort-spAge_with-slider'),
                        dcc.Slider(
                            20,
                            mb['Age'].max(),
                            step = 10,
                            value = 20,
                            id = 'Age_slider'
                        ),
                        dcc.RadioItems(mb['Ethnicity'].unique(),
                                       id = 'Ethnicity_radio')])])

@callback(
        Output('spComfort-spAge_with-slider', 'figure'),
        Input('Age_slider', 'value'),
        Input('Ethnicity_radio', 'value')
)
def update_figure(selected_age, selected_ethnicity):
    mb_filtered = mb[(mb['Age'] < selected_age) & (mb['Age'] > selected_age - 10) & (mb['Ethnicity'] == selected_ethnicity)]

    fig = px.scatter(mb_filtered, x = 'spAge', y = 'spComfort', color = 'Sex_c')
    return fig

if __name__ == '__main__':
    app6.run(debug = True)
```

```python
app7 = Dash()

app7.layout = html.Div([html.H1('Hello'),
                        html.H3('My App is great'),
                        html.Div([
                        dcc.Graph(id = 'spComfort-spAge_with-slider'),
                        dcc.Slider(
                            20,
                            mb['Age'].max(),
                            step = 10,
                            value = 20,
                            id = 'Age_slider'
                        ),
                        dcc.RadioItems(mb['Ethnicity'].unique(),
                                       id = 'Ethnicity_radio')])])

@callback(
        Output('spComfort-spAge_with-slider', 'figure'),
        Input('Age_slider', 'value'),
        Input('Ethnicity_radio', 'value')
)
def update_figure(selected_age, selected_ethnicity):
    mb_filtered = mb[(mb['Age'] < selected_age) & (mb['Age'] > selected_age - 10) & (mb['Ethnicity'] == selected_ethnicity)]

    fig = px.scatter(mb_filtered, x = 'spAge', y = 'spComfort', color = 'Sex_c', title = 'Age by Comfort for {} individuals'.format(selected_ethnicity))
    return fig

if __name__ == '__main__':
    app7.run(debug = True)
```
