```python
import pandas as pd
import plotly.express as px
import dash
from dash import Dash, html, dcc, Output, Input, callback
```

```python
mb = pd.read_csv('mobile-use.csv')
mb.columns
```

```python
app1 = Dash()
app1.layout = html.Div([
    html.H1("David's Awesome App"),
    html.H2("Some subheading"),
    dcc.Graph(id = 'spComfort-spAge_figure'),
    dcc.Slider(id = 'Age_slider')
])




if __name__ == '__main__':
    app1.run(debug = True)
```

```python
app2 = Dash()
app2.layout = html.Div([
    html.H1("David's Awesome App"),
    html.H2("Some subheading"),
    dcc.Graph(id = 'spComfort-spAge_figure'),
    dcc.Slider(mb['Age'].min(),
               mb['Age'].max(),
               step = 1,
               value = mb['Age'].min(),
               id = 'Age_slider')
])

@callback(
    Output('spComfort-spAge_figure', 'figure'),
    Input('Age_slider', 'value')
)
def update_fig(age_select):
    fig = px.scatter(mb, x = 'spAge', y = 'spComfort', color = 'Sex')
    return fig

if __name__ == '__main__':
    app2.run(debug = True)
```

```python
app3 = Dash()
app3.layout = html.Div([
    html.H1("David's Awesome App"),
    html.H2("Some subheading"),
    dcc.Graph(id = 'spComfort-spAge_figure'),
    dcc.Slider(mb['Age'].min(),
               mb['Age'].max(),
               step = 1,
               value = mb['Age'].min(),
               id = 'Age_slider')
])

@callback(
    Output('spComfort-spAge_figure', 'figure'),
    Input('Age_slider', 'value')
)
def update_fig(age_select):
    filtered_mb = mb[mb['Age'] == age_select]
    fig = px.scatter(filtered_mb, x = 'spAge', y = 'spComfort', color = 'Sex')
    return fig

if __name__ == '__main__':
    app3.run(debug = True)
```

```python
app4 = Dash()
app4.layout = html.Div([
    html.H1("David's Awesome App"),
    html.H2("Some subheading"),
    dcc.Graph(id = 'spComfort-spAge_figure'),
    dcc.Slider(20,
               90,
               step = 10,
               value = 50,
               id = 'Age_slider')
])

@callback(
    Output('spComfort-spAge_figure', 'figure'),
    Input('Age_slider', 'value')
)
def update_fig(age_select):
    filtered_mb = mb[(mb['Age'] < age_select + 10) & (mb['Age'] >= age_select)]
    fig = px.scatter(filtered_mb, x = 'spAge', y = 'spComfort', color = 'Sex_c',
                    title = 'SP Comfort by SP Age for people {}-{} years old'.format(age_select, age_select+10))
    # f'SP Comfort by SP Age for people {age_select}-{age_select+10} years old'
    return fig

if __name__ == '__main__':
    app4.run(debug = True)
```

```python
app5 = Dash()
app5.layout = html.Div([
    html.H1("David's Awesome App"),
    html.H2("Some subheading"),
    dcc.Graph(id = 'spComfort-spAge_figure'),
    dcc.Slider(20,
               90,
               step = 10,
               value = 50,
               id = 'Age_slider'),
    dcc.RadioItems(mb['Ethnicity'].unique(),
                   value = 'Black or African American',
                  id = 'Ethnicity_radio')
])

@callback(
    Output('spComfort-spAge_figure', 'figure'),
    Input('Age_slider', 'value'),
    Input('Ethnicity_radio', 'value')
)
def update_fig(age_select, ethnicity_select):
    filtered_mb = mb[(mb['Age'] < age_select + 10) & (mb['Age'] >= age_select) & (mb['Ethnicity'] == ethnicity_select)]
    fig = px.scatter(filtered_mb, x = 'spAge', y = 'spComfort', color = 'Sex_c',
                    title = 'SP Comfort by SP Age for people {}-{} years old'.format(age_select, age_select+10))
    return fig

if __name__ == '__main__':
    app5.run(debug = True)
```
